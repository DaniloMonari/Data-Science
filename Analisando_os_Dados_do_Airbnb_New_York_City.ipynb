## **Sobre Nova York**


**Distritos**

Nova York é dividida em cinco regiões principais, conhecidas como boroughs ou distritos. Cada distrito tem sua própria identidade e características únicas. A seguir, descrevo brevemente cada um deles:

**Manhattan:** Localizado no centro de Nova York, Manhattan é o coração da cidade e abriga muitas das atrações mais famosas. Possui cerca de 59,1 quilometros quadrados, os principais pontos de interesse entre os turistas são: Times Square, Central Park, Empire State Building, Estátua da Liberdade e Ilha Ellis, Museu Metropolitano de Arte (Met), Memorial do 11 de Setembro, Museu de Arte Moderna (MoMA), High Line, Broadway, Quinta Avenida e Chinatown.

**Brooklyn:** Localizado ao sul de Manhattan, o Brooklyn é conhecido por sua vibração artística, bairros históricos e diversidade cultural. Abriga o Brooklyn Bridge, o Prospect Park, o Brooklyn Museum e o bairro movimentado de Williamsburg, conhecido por suas galerias de arte, restaurantes descolados e vida noturna agitada.

**Queens:** Situado a leste de Manhattan, o Queens é o maior distrito de Nova York em termos de área geográfica. É uma área culturalmente diversa e étnica, com uma mistura de bairros residenciais e comerciais. O Queens abriga o Aeroporto JFK, o Flushing Meadows-Corona Park (onde ocorreram as Feiras Mundiais de 1939 e 1964), o Citi Field (estádio dos New York Mets) e o bairro de Astoria, conhecido por seus restaurantes étnicos e cenas artísticas.

**Bronx:** Localizado ao norte de Manhattan, o Bronx é conhecido por ser o lar do Yankee Stadium, onde o time de beisebol New York Yankees joga. O distrito também abriga o Jardim Zoológico do Bronx, o Jardim Botânico de Nova York e a famosa avenida Grand Concourse, com sua arquitetura Art Deco.

**Staten Island:** Situado ao sul de Manhattan, Staten Island é um distrito mais tranquilo e suburbano, com uma conexão com Manhattan através da Staten Island Ferry. A ilha é conhecida pelo Parque Histórico Nacional de Staten Island, o Jardim Botânico de Snug Harbor e pela área costeira pitoresca.

Esses cinco distritos juntos formam a cidade de Nova York, cada um com sua própria personalidade e atrações distintas. Cada distrito possui uma ampla variedade de bairros e comunidades que contribuem para a rica diversidade cultural da cidade

## Obtenção dos Dados


Os dados utilizados neste estudo foram obtidos no site [Inside Airbnb](http://insideairbnb.com/explore/). O arquivo utilizado, denominado "listings.csv", foi coletado em 2 de junho de 2023 e contém informações referentes à data de 6 de março de 2023. Para realizar a análise inicial, optou-se pelo uso do arquivo resumido em vez do completo, uma vez que se considerou que as informações resumidas eram suficientes para atender aos objetivos propostos para este estudo.

```
# importar os pacotes necessarios
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```
# importar o arquivo listings.csv para um DataFrame
df = pd.read_csv("http://data.insideairbnb.com/united-states/ny/new-york-city/2023-03-06/visualisations/listings.csv")
```

#### Boxplot para price e minimum_nights:

```
df.price.plot(kind='box', vert=False, figsize=(15, 3))
plt.show()

print("\nprice: valores acima de 1500:")
print("{} entradas".format(len(df[df.price > 1500])))
print("{:.4f}%".format((len(df[df.price > 1500]) / df.shape[0])*100))
```

```
df.minimum_nights.plot(kind='box', vert=False, figsize=(15, 3))
plt.show()

print("\minimum_nights: valores acime de 90:")
print("{} entradas".format(len(df[df.minimum_nights > 90])))
print("{:.4f}%".format((len(df[df.minimum_nights > 90]) / df.shape[0])*100))
```

#### Remover os outliers e plotar o histograma novamente

```
# remover os *outliers* em um novo DataFrame
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 1500].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 90].index, axis=0, inplace=True)

# remover as colunas Neighbourhood_group e license, pois estão sem dados
df_clean.drop(['license'],axis=1, inplace=True)

# plotar o histograma para as variáveis numéricas
df_clean.hist(bins=15, figsize=(15,10));
```

### **Q4. Qual a correlação existente entre as variáveis**

```
# criar uma matriz de correlação
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()
# mostrar a matriz de correlação
display(corr)
```

```
# plotar um heatmap a partir das correlações

sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

## Conclusões

Com base na análise dos dados das propriedades disponíveis em New York no Airbnb, podemos chegar a algumas conclusões importantes sobre os preços, média do número de diárias e os grupos de bairros mais baratos e mais caros. Essas informações podem ser valiosas para proprietários de imóveis e viajantes que desejam tomar decisões informadas sobre alocar ou alugar um imóvel.

Em relação aos preços, identificamos que eles variam significativamente de acordo com diversos fatores, como localização, tipo de quarto e quantidade minima de noites. Os tipos de acomodações mais locadas são 
 **56% Entire home/apt** e **41% Private room**, já os bairros mais próximos de áreas turísticas ou com boa infraestrutura tendem a ter preços mais altos, enquanto bairros afastados ou menos conhecidos geralmente oferecem acomodações mais acessíveis. O distrito de bairro mais caro é Manhattan, com uma média de **225.55 USD** por noite, já o mais barato é o Bronx com um preço médio de **109 USD** por noite. 

Quanto ao número de diárias, observamos uma variação considerável entre as propriedades analisads. Alguns proprietários preferem alugar seus imóveis apenas por curtos períodos, enquanto outros optam por oferecer opções de longa duração. A média do mínimo de noites é de **18**, identificamos também que aproximadamente **42%** das locações exigem o mínimo de 30 noites, e **50%** realizam locações abaixo de 30 noites, que merece uma atenção maior devido a nova lei. 

É importante ressaltar que essas conclusões são baseadas na análise de dados coletadas no dia 2 de junho de 2023. Os preços, número de diárias e a dinâmica dos bairros podem sofrer alterações ao longo do tempo devido a diversos fatores, como demanda, oferta e flutuações econômicas.
